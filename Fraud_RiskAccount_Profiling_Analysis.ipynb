## Key Insights

Based on the analysis performed so far, here are some key insights:

*   **Data Overview:** The dataset contains 50,000 entries and 23 columns, with no missing values, making it clean for analysis.
*   **Target Variable Distribution:** There's a class imbalance in the `is_fraudster` column, with approximately 36,664 non-fraudulent accounts (0.0) and 13,336 fraudulent accounts (1.0). This imbalance is important to consider for model training.
*   **Risk Score and Fraud:** The boxplot of `risk_score` vs. `is_fraudster` suggests that fraudulent accounts tend to have higher risk scores on average.
*   **Transaction Behavior:** The boxplots for `avg_txn_amount` and `pct_foreign` in relation to `is_fraudster` indicate potential differences in transaction patterns between fraudulent and non-fraudulent accounts. For instance, fraudulent accounts might exhibit different average transaction values or a higher proportion of foreign transactions.
*   **2FA Usage:** The countplot shows how `has_2fa` is distributed across fraudulent and non-fraudulent accounts. This can reveal if the presence or absence of 2FA significantly impacts fraud rates.
*   **Feature Correlations:** The heatmap provides an overview of correlations between numerical features, which can help in understanding feature relationships and for feature selection in a predictive model.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = df.drop(['is_fraudster', 'account_id', 'home_country'], axis=1)
y = df['is_fraudster']

X = pd.get_dummies(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

print("Accuracy:", model.score(X_test, y_test))
```
